### Merge

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq
```

```python
der
```

```python
izq.merge(der)
```

```python
izq.merge(der,on= 'key')
```

```python
izq2 = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der2 = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq2.merge(der2,on= 'key')
```

```python
izq2.merge(der2, left_on='key', right_on='key_2')
```

```python
izq2 = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der2 = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2',np.nan],
'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
der2
```

```python
izq2.merge(der2, left_on='key', right_on='key_2', how='left')
```

```python
izq2.merge(der2, left_on='key', right_on='key_2', how='right')
```

```python
izq2.merge(der2, left_on='key', right_on='key_2', how='outer')
```

### Join

```python
izq = pd.DataFrame({'A': ['A0','A1','A2'],
'B':['B0','B1','B2']},
index=['k0','k1','k2'])

der =pd.DataFrame({'C': ['C0','C1','C2'],
'D':['D0','D1','D2']},
index=['k0','k2','k3'])
```

```python
izq.join(der)
```

```python
izq.join(der, how='inner')
```

```python
izq.join(der, how='left')
```

```python
izq.join(der, how='right')
```

```python
izq.join(der, how='outer')
```
